```python
import pandas as pd
import numpy as np
import re
import math
```

```python
df_WebMD = pd.read_csv('/Users/siyuchen/Desktop/dataset/WebMD_10_8.csv', dtype=object)
df_ClevelandClinic = pd.read_csv('/Users/siyuchen/Desktop/dataset/ClevelandClinic_10_8.csv', dtype=object)
df_MayoClinic = pd.read_csv('/Users/siyuchen/Desktop/dataset/MayoClinic_10_8.csv', dtype=object)

df_ls = [df_WebMD, df_ClevelandClinic, df_MayoClinic]
df_title = ['df_WebMD.csv', 'df_ClevelandClinic.csv', 'df_MayoClinic.csv']
```

```python
def process(df):
    text = []
    url_twitter = []
    check = []

    content_index = 0
    for t in df['content'].isnull():
        if t == True: check.append(content_index)
        content_index = content_index+1

    j = 0
    for i in df['content']:
        #print(j)
        if j in check: 
            text.append(np.nan)
            url_twitter.append(np.nan)
        else: 
            text.append(re.split("http",i)[0])
            if len(re.split("http",i)) == 2: 
                url_twitter.append(re.split("http",i)[1])
        j=j+1

        #     else url_twitter.append(None)
    df['text'] = text

    media = []
    df_media = df[['image', 'articleImage', 'video']]
    df_media_check = df_media.isnull()
    for index, row in df_media_check.iterrows():
        if row.tolist()[0] == False or row.tolist()[1] == False: media.append(1)
        if row.tolist()[0] == True and row.tolist()[1] == True and row.tolist()[2] == True: media.append(1)
        if row.tolist()[0] == True and row.tolist()[1] == True and row.tolist()[2] == False: media.append(2)        
    df['media'] = media
    df = df[['username', 'TwitterID', 'DateTime', 'likes', 'retweet', 'replies', 'text', 'media']]
    
    return df
```

```python
new_df = []
for j in df_ls:
    result = process(j)
    new_df.append(result)
```
